## Z Test for Two samples

```python
import numpy as np
from scipy import stats
```

### Question: Two-sample Z-test (two-tailed)

```python
sample_mean_A = 48
sample_mean_B = 50
n1 = 150
n2 = 160
population_std1 = 3
population_std2 = 4
significance_value = 0.05
```

```python
z_stats = (sample_mean_A - sample_mean_B)/np.sqrt((np.square(population_std1)/n1)+(np.square(population_std2)/n2))
z_stats
```

```python
p_value_two_tailed = 2*(1-stats.norm.cdf(abs(z_stats)))
```

```python
if p_value_two_tailed > significance_value:
    print("I will accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```

### Two-sample Z-test (right-tailed)

```python
sample_mean_A = 82
sample_mean_B = 78
n1 = 64
n2 = 100
population_std1 = 6
population_std2 = 5
significance_value = 0.05
```

```python
z_stats = (sample_mean_A - sample_mean_B)/np.sqrt((np.square(population_std1)/n1)+(np.square(population_std2)/n2))
z_stats
```

```python
p_value_right_tailed = 1 - stats.norm.cdf(z_stats)
p_value_right_tailed
```

```python
if p_value_two_tailed > significance_value:
    print("I will accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```
