## Z Test of Sample

```python
import numpy as np
from scipy.stats import norm
```

```python
sample = np.array([172,174,168,169,171,173,175,170,169,172])
```

```python
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean - population_mean)/(population_std/np.sqrt(n))
print(z_score)
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(p_value)
```

```python
significance_value = 0.05    #alpha value
```

```python
if p_value>significance_value:
    print("I will accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```
