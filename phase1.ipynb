## How can we remove the outliers from the array

```python
import numpy as np
import seaborn as sns
```

```python
arr = np.array([10, 12, 13, 15, 16, 18, 19, 20, 100, -30])
arr
```

#### For removing outliers, we will use IQR

```python
Q1 = np.percentile(arr, 25)
Q2 = np.percentile(arr, 50)
Q3 = np.percentile(arr, 75)

IQR = Q3 - Q1


Upper_fence = Q3 + 1.5 * IQR
Lower_fence = Q1 - 1.5 * IQR


new_arr = arr[(arr >= Lower_fence) & (arr <= Upper_fence)]
print("So finally, we removed all the outliers from the array = ",new_arr)
```

### Now we will make its box plot

```python
sns.boxplot(x=new_arr) # Without outlier
sns.boxplot(x=arr) # With outlier
```
