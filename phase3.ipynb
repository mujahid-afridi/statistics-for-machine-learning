## T-Test of Sample

```python
import numpy as np
from scipy import stats
```

```python
sample = np.array([172,174,168,169,171,173,175,170,169,172])
sample
```

```python
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)
mean_population = 170    #Given
```

```python
t_stats = (sample_mean - mean_population)/(sample_std/np.sqrt(n))
t_stats
```

```python
p_value = 2*(1 - stats.t.cdf(abs(t_stats), df=n-1))
p_value
```

```python
significance_value = 0.05    #alpha value
```

```python
if p_value>significance_value:
    print("I wil accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```
