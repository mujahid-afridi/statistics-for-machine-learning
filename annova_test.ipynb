## ANOVA Test(analysis of variance) or F test

```python
import numpy as np
import seaborn as sns
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df = df[["age", "pclass"]].dropna()
```

```python
df
```

```python
df["pclass"].unique()
```

```python
class1 = df[df["pclass"]==1]["age"]
class2 = df[df["pclass"]==2]["age"]
class3 = df[df["pclass"]==3]["age"]
```

```python
f_stats, p_value = f_oneway(class1,class2, class3)
f_stats
```

```python
significance_value = 0.05
```

```python
if p_value > significance_value:
    print("we accept the null hypothesis. And there is no signifiant difference.")
else:
    print("We reject the null hypothesis , and there is significant difference in at least one passenger class.")
```
