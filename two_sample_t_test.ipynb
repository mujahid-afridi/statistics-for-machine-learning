##### Can find T test and p value directly by using builtin methods

```python
t_test, p_value = stats.ttest_ind(group_A, group_B ,equal_var=False)
```

```python
significance_value = 0.05      #alpha value
```

```python
if p_value > significance_value:
    print("I wil accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```

### Two samples T test (right tail)

```python
program_A = [3.1, 2.9, 3.5, 3.2, 3.0, 3.3, 2.8, 3.4]
program_B = [2.4, 2.6, 2.5, 2.3, 2.7, 2.5, 2.8, 2.6]
```

```python
t_stats, p_value = stats.ttest_ind(program_A, program_B, equal_var=False)
p_value
```

```python
p_right = p_value/2 if t_stats>0 else 1-(p_value/2)
p_right
```

```python
significance_value = 0.05      #alpha value
```

```python
if p_right > significance_value:
    print("I wil accept the null hypothesis")
else:
    print("I will reject the null hypothesis")
```
